## LOAN APPROVAL PREDICTION

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("loan_prediction.csv")
```

```python
print(df.columns)
```

## 1.LOAN STATUS DISTRIBUTION

```python
loan_status = df['Loan_Status'].value_counts()

print(loan_status)

plt.figure(figsize=(6,5))

plt.bar(loan_status.index,
        loan_status.values,
        color=['lightgreen', 'lightcoral'])

plt.title("Loan Status Distribution")
plt.xlabel("Loan Status")
plt.ylabel("Number of Applicants")

# Display values on top of bars
for i, value in enumerate(loan_status.values):
    plt.text(i, value + 5, str(value), ha='center')

plt.show()
```

```python
plt.figure(figsize=(6,6))

plt.pie(loan_status.values,
        labels=['Approved', 'Rejected'],
        autopct='%1.1f%%',
        colors=['mediumpurple', 'lightpink'],
        startangle=90)

plt.title("Loan Approval Percentage")
plt.axis('equal')

plt.show()
```

## 2.GENDER VS LOAN APPROVAL

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


gender_status = pd.crosstab(df['Gender'], df['Loan_Status'])

print(gender_status)


x = np.arange(len(gender_status.index))
width = 0.35

plt.figure(figsize=(7,5))


plt.bar(x - width/2,
        gender_status['Y'],
        width,
        label='Approved',
        color='lightgreen')


plt.bar(x + width/2,
        gender_status['N'],
        width,
        label='Rejected',
        color='lightcoral')


plt.xticks(x, gender_status.index)
plt.xlabel("Gender")
plt.ylabel("Number of Applicants")
plt.title("Gender vs Loan Approval")
plt.legend()


for i, value in enumerate(gender_status['Y']):
    plt.text(i - width/2, value + 2,
             str(value), ha='center')

for i, value in enumerate(gender_status['N']):
    plt.text(i + width/2, value + 2,
             str(value), ha='center')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


gender_percentage = pd.crosstab(
    df['Gender'],
    df['Loan_Status'],
    normalize='index'
) * 100

plt.figure(figsize=(10,5))


plt.subplot(1,2,1)
plt.pie(
    gender_percentage.loc['Male'],
    labels=['Approved', 'Rejected'],
    autopct='%1.1f%%',
    colors=['skyblue', 'orange'],
    startangle=90,
    explode=(0.05, 0)
)
plt.title("Male Loan Approval Percentage")


plt.subplot(1,2,2)
plt.pie(
    gender_percentage.loc['Female'],
    labels=['Approved', 'Rejected'],
    autopct='%1.1f%%',
   colors=['skyblue', 'orange'],
    startangle=90,
    explode=(0.05, 0)
)
plt.title("Female Loan Approval Percentage")

plt.suptitle("Gender-wise Loan Approval Percentage")
plt.tight_layout()
plt.show()
```

## 6.APPLICANT INCOME DISTRIBUTION

```python

import matplotlib.pyplot as plt

income_bins = pd.cut(df['ApplicantIncome'],
                     bins=[0,3000,6000,9000,20000],
                     labels=['0-3000','3001-6000','6001-9000','9001+'])

income_count = income_bins.value_counts().sort_index()

plt.figure(figsize=(7,5))
plt.bar(income_count.index.astype(str), income_count.values,
        color=['skyblue','green','red','orange'])

plt.xlabel("Applicant Income Range")
plt.ylabel("Number of Applicants")
plt.title("Applicant Income Distribution")

for i, value in enumerate(income_count.values):
    plt.text(i, value+2, str(value), ha='center')

plt.show()
```

## 7.LOAN AMOUNT DISTRIBUTION

```python
import pandas as pd
import matplotlib.pyplot as plt


loan_range = pd.cut(df['LoanAmount'],
                    bins=[0, 100, 200, 300, 400, df['LoanAmount'].max()],
                    labels=['0-100', '101-200', '201-300', '301-400', '400+'])


loan_count = loan_range.value_counts().sort_index()

plt.figure(figsize=(7,5))
plt.bar(loan_count.index.astype(str),
        loan_count.values,
        color=['green','red','orange','yellow','blue'])

plt.xlabel("Loan Amount Range")
plt.ylabel("Number of Applicants")
plt.title("Loan Amount Distribution")

# Display counts
for i, value in enumerate(loan_count.values):
    plt.text(i, value + 2, str(value), ha='center')

plt.tight_layout()
plt.show()
```

## 11.Property Area vs Loan Status (Urban, Rural, Semiurban) Stacked Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

property_status = pd.crosstab(df['Property_Area'], df['Loan_Status'])

print(property_status)


property_status.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5),
    color=['powderblue', 'palegreen']
)

plt.title("Property Area vs Loan Status")
plt.xlabel("Property Area")
plt.ylabel("Number of Applicants")
plt.legend(['Rejected', 'Approved'])

for i in range(len(property_status)):
    rejected = property_status.iloc[i]['N']
    approved = property_status.iloc[i]['Y']

    plt.text(i, rejected/2,
             str(rejected),
             ha='center')

    plt.text(i,
             rejected + approved/2,
             str(approved),
             ha='center')

plt.tight_layout()
plt.show()
```
